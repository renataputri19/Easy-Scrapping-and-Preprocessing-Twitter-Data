```python
import pandas as pd
data1 = pd.read_csv("twitter#crypto.csv")
data2 = pd.read_csv("twitter#Kripto.csv")
data3 = pd.read_csv("twitterKripto.csv")
data4 = pd.read_csv("twitterKriptomui.csv")
```

```python
frame = [data1, data2, data3, data4]
data_keseluruhan = pd.concat(frame).reset_index()
data_keseluruhan.to_csv('data_keseluruhan.csv')
```

```python
data_keseluruhan = data_keseluruhan.drop(['index', 'id','conversation_id','mentions','urls','photos'
         ,'replies_count','retweets_count','likes_count','hashtags','cashtags',
         'link','retweet','quote_url','video','thumbnail','user_rt_id',
        'user_rt','retweet_id','reply_to','trans_src','trans_dest',
            'translate','retweet_date','source','geo','near','language','user_id',
                      'timezone','name','place'], axis = 1)
```
